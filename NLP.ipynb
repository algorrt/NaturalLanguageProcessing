```python
import string
import re
import nltk
from nltk.tokenize import TweetTokenizer, word_tokenize

# tweet = df['Sound Bite Text'][2]
# print(tweet)
allPunc = ["\"","\\","'",".",">",":","<",",","?","}","{","[","]","/","|","+","=","_","-",")","(","~","“","”","…","'s"]
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # remove numbers
#     text_nonum = re.sub(r'\d+', '', text)
    # remove punctuations and convert characters to lower case
    for i in allPunc:
        text = text.replace(i,"")
    text = text.lower()
    
    tt = TweetTokenizer()
    text = tt.tokenize(text)
    
    for i in stop_words:
        text = [j for j in text if j!=i]
    
#     text_nopunct = "".join([char.lower() for char in text_nonum if char not in string.punctuation]) 
    # substitute multiple whitespace with single whitespace
    # Also, removes leading and trailing whitespaces
#     text_no_doublespace = re.sub('\s+', ' ', text_nopunct).strip()
    return text


for i in range(len(df['Sound Bite Text'])):
    tweet = df['Sound Bite Text'][i]
#     print(tweet)
    if str(type(tweet)) == "<class 'str'>":
        cur = clean_text(tweet)
        print(cur)
        df['Sound Bite Text'][i] = cur
    else:
#         print(i)
        df['Sound Bite Text'][i] = ""
        
    
# print(cleaned_tweet)
# tt = TweetTokenizer()
# print(tt.tokenize(cleaned_tweet))
```
